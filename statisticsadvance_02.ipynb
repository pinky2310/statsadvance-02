**1.What is hypothesis testing in statistics?**
- Hypothesis testing is a statistical method used to make decisions about a population based on sample data. It involves formulating two competing hypotheses: the null hypothesis (H0) and the alternative hypothesis (H1 or Ha). The goal is to determine whether there is enough evidence in the sample data to reject the null hypothesis in favor of the alternative hypothesis

---



**2.What is the null hypothesis, and how does it differ from the alternative hypothesis?**
- The null hypothesis (H0) is a statement that there is no effect or no difference, and it serves as the default assumption that any observed effect in the data is due to random chance. The alternative hypothesis (H1 or Ha) is the statement that there is an effect or a difference. The key difference is that the null hypothesis represents a position of no change, while the alternative hypothesis represents a position of change or effect.

---



**3.What is the significance level in hypothesis testing, and why is it important?**
- The significance level (denoted as α) is the threshold for determining whether to reject the null hypothesis. It is the probability of making a Type I error, which occurs when the null hypothesis is incorrectly rejected when it is actually true. Common significance levels are 0.05, 0.01, and 0.10. The significance level is important because it helps researchers control the risk of making incorrect conclusions based on their data.

---



**4.What does a P-value represent in hypothesis testing?**
 - The P-value is the probability of obtaining test results at least as extreme as the observed results, assuming that the null hypothesis is true. It quantifies the strength of the evidence against the null hypothesis. A smaller P-value indicates stronger evidence against H0.

---



**5.How do you interpret the P-value in hypothesis testing?**
- To interpret the P-value, compare it to the significance level (α). If the P-value is less than or equal to α, you reject the null hypothesis, suggesting that the observed data is statistically significant. If the P-value is greater than α, you fail to reject the null hypothesis, indicating insufficient evidence to support the alternative hypothesis.

---



**6.What are Type 1 and Type 2 errors in hypothesis testing?**
Type I Error (α): This error occurs when the null hypothesis is rejected when it is actually true. It is a false positive.
Type II Error (β): This error occurs when the null hypothesis is not rejected when it is actually false. It is a false negative.

---



**7.What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**
- One-tailed test: This test evaluates the direction of the effect (either greater than or less than a certain value). It tests for the possibility of the relationship in one direction.
- Two-tailed test: This test evaluates the effect in both directions (either greater than or less than a certain value). It tests for the possibility of the relationship in both directions.

---



**8.What is the Z-test, and when is it used in hypothesis testing?**
- The Z-test is a statistical test used to determine whether there is a significant difference between the means of two groups or between a sample mean and a population mean when the population variance is known. It is typically used when the sample size is large (n > 30) or when the population is normally distributed.

---



**9.How do you calculate the Z-score, and what does it represent in hypothesis testing?**
- The Z-score is calculated using the formula:

[ Z = \frac{(X - \mu)}{\sigma} ]

where:

(X) is the sample mean,

(\mu) is the population mean,

(\sigma) is the population standard deviation.

The Z-score represents the number of standard deviations a data point is from the mean. In hypothesis testing, it helps determine how far the sample mean is from the null hypothesis mean in terms of standard deviations.

---



**10.What is the T-distribution, and when should it be used instead of the normal distribution?**
- The T-distribution is a probability distribution that is used when the sample size is small (typically n < 30) and/or when the population standard deviation is unknown. It is similar to the normal distribution but has heavier tails, which accounts for the increased variability expected in smaller samples. The T-distribution is used in hypothesis testing when estimating population parameters from small samples.

---



**11.What is the difference between a Z-test and a T-test?**
- The primary difference between a Z-test and a T-test lies in the sample size and whether the population standard deviation is known:

   - Z-test: Used when the sample size is large (n > 30) or when the population standard deviation is known. It assumes that the sampling distribution of the sample mean is normally distributed.
   - T-test: Used when the sample size is small (n < 30) and/or when the population standard deviation is unknown. The T-test accounts for the additional uncertainty in estimating the population standard deviation from a small sample.

---



**12.What is the T-test, and how is it used in hypothesis testing?**
  - The T-test is a statistical test used to determine if there is a significant difference between the means of two groups or between a sample mean and a population mean. It is commonly used in hypothesis testing when the sample size is small and the population standard deviation is unknown. The T-test calculates a T-statistic, which is then compared to a critical value from the T-distribution to determine whether to reject the null hypothesis.

---



**13.What is the relationship between Z-test and T-test in hypothesis testing?**
- Both the Z-test and T-test are used to test hypotheses about population means. The relationship between them is that as the sample size increases, the T-distribution approaches the normal distribution. Therefore, for large samples, the results of a T-test and a Z-test will converge, and they will yield similar conclusions. However, for small samples, the T-test is preferred due to its ability to account for the additional variability.

**14.What is a confidence interval, and how is it used to interpret statistical results?**
- A confidence interval is a range of values, derived from sample data, that is likely to contain the true population parameter (e.g., mean) with a specified level of confidence (e.g., 95% or 99%). It provides an estimate of the uncertainty around the sample statistic. For example, a 95% confidence interval means that if the same sampling procedure were repeated many times, approximately 95% of the calculated intervals would contain the true population parameter.

It is calculated based on the standard error and the critical value from the Z or T distribution corresponding to the desired confidence level. The margin of error affects the width of the confidence interval: a larger margin of error results in a wider confidence interval, indicating more uncertainty about the estimate, while a smaller margin of error results in a narrower confidence interval.

**16.How is Bayes' Theorem used in statistics, and what is its significance?**

This is particularly useful in fields such as Bayesian statistics, machine learning, and decision-making under uncertainty.

It is used primarily in tests of independence and goodness of fit. The distribution is characterized by its degrees of freedom, which are determined by the number of categories or groups being analyzed.

**18.What is the Chi-square goodness of fit test, and how is it applied?**
- The Chi-square goodness of fit test is a statistical test used to determine whether the observed frequencies of a categorical variable differ significantly from the expected frequencies. It is applied by calculating the Chi-square statistic, which compares the observed counts to the expected counts under the null hypothesis. If the Chi-square statistic exceeds a critical value from the Chi-square distribution, the null hypothesis is rejected, indicating that the observed distribution is significantly different from the expected distribution.

---



**19.What is the F-distribution, and when is it used in hypothesis testing?**
The F-distribution is a probability distribution that arises in the context of variance analysis. It is used primarily in hypothesis testing when comparing the variances of two or more groups. The F-distribution is characterized by two sets of degrees of freedom: one for the numerator (between-group variance) and one for the denominator (within-group variance). It is commonly used in ANOVA (Analysis of Variance) tests.

---



**20.What is an ANOVA test, and what are its assumptions?**
- ANOVA (Analysis of Variance) is a statistical method used to compare the means of three or more groups to determine if at least one group mean is significantly different from the others. It helps to assess whether the variability among group means is greater than the variability within the groups, which would suggest that at least one group is different.

**Assumptions of ANOVA:**

Independence: The samples must be independent of each other. This means that the data collected from one group should not influence the data collected from another group.
Normality: The data in each group should be approximately normally distributed. This assumption can be checked using normality tests (e.g., Shapiro-Wilk test) or visual methods (e.g., Q-Q plots).
Homogeneity of Variances: The variances among the groups should be approximately equal. This can be tested using Levene's test or Bartlett's test. If this assumption is violated, a different statistical method may be more appropriate.

---




**21.What are the different types of ANOVA tests?**
- There are several types of ANOVA tests, including:

   - One-Way ANOVA: Compares the means of three or more independent groups based on one independent variable (factor). For example, comparing the test scores of students from different teaching methods.

   - Two-Way ANOVA: Compares the means of groups based on two independent variables (factors) and can also assess the interaction between the two factors. For example, examining the effects of teaching method and gender on student performance.

   - Repeated Measures ANOVA: Used when the same subjects are measured multiple times under different conditions. For example, measuring the same group of patients' blood pressure at different times after treatment.

  - MANOVA (Multivariate Analysis of Variance): Extends ANOVA when there are two or more dependent variables. It assesses whether the mean differences among groups on a combination of dependent variables are significant.

---



**22.What is the F-test, and how does it relate to hypothesis testing?**
- The F-test is a statistical test used to compare the variances of two or more groups. In the context of ANOVA, the F-test is used to determine whether the variability between group means is significantly greater than the variability within the groups.
-**Relation to Hypothesis Testing:**

-In ANOVA, the null hypothesis (H0) states that all group means are equal (i.e., there is no effect of the independent variable on the dependent variable).

- The alternative hypothesis (H1) states that at least one group mean is different from the others.

- The F-statistic is calculated as the ratio of the variance between the groups to the variance within the groups. A larger F-statistic indicates that the group means are more spread out than would be expected by chance.

- If the calculated F-statistic exceeds the critical value from the F-distribution (based on the chosen significance level and degrees of freedom), the null hypothesis is rejected, suggesting that there is a significant difference among the group means.

**1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.**

```python
import numpy as np
import scipy.stats as stats

# Function to perform a one-sample Z-test
def z_test(sample, population_mean, population_std):
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test
    return z_score, p_value

# Simulate random data
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=10, size=30)  # Sample mean = 100, std = 10
population_mean = 95  # Known population mean
population_std = 10   # Known population std

# Perform Z-test
z_score, p_value = z_test(sample_data, population_mean, population_std)

# Interpret results
alpha = 0.05
print(f"Z-Score: {z_score:.2f}, P-Value: {p_value:.4f}")
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the sample mean and the population mean.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the sample mean and the population mean.")
```

**5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.**

---



```python
def visualize_errors(alpha, beta):
    z_critical = stats.norm.ppf(1 - alpha/2)
    z_power = stats.norm.ppf(1 - beta)

    # Plotting
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)

    plt.plot(x, y, label='Standard Normal Distribution')
    plt.fill_between(x, y, where=(x < -z_critical) | (x > z_critical), color='red', alpha=0.5, label='Type I Error Region')
    plt.fill_between(x, y, where=(x > z_power), color='blue', alpha=0.5, label='Type II Error Region')
    plt.axvline(z_critical, color='red', linestyle='--', label='Critical Value')
    plt.axvline(z_power, color='blue', linestyle='--', label='Power Threshold')
    plt.title('Type I and Type II Errors')
    plt.xlabel('Z-value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.show()

visualize_errors(0.05, 0.2)
```

**9.Write a Python Function to Calculate the Confidence Interval for a Sample Mean and explain its significance..**

---



```python
def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    data (array-like): Sample data
    confidence (float): Confidence level (default is 0.95)

    Returns:
    tuple: Lower and upper bounds of the confidence interval
    """
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean
    h = std_err * stats.t.ppf((1 + confidence) / 2, n - 1)  # Margin of error
    return mean - h, mean + h

# Example usage
sample_data = np.random.normal(loc=50, scale=10, size=100)
ci = confidence_interval(sample_data)
print(f"Confidence Interval: {ci[0]:.2f} to {ci[1]:.2f}")
```

**Significance of Confidence Interval:**

The confidence interval provides a range of values within which we can be reasonably certain that the true population mean lies. A wider interval indicates more uncertainty, while a narrower interval suggests more precision in our estimate. The confidence level (e.g., 95%) indicates the probability that the interval contains the true mean if we were to take many samples.



**10.Write a Python program to calculate the margin of error for a given confidence level using sample data.**

---




```python
import numpy as np
import scipy.stats as stats

def margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for a given confidence level using sample data.

    Parameters:
    data (array-like): Sample data
    confidence (float): Confidence level (default is 0.95)

    Returns:
    float: Margin of error
    """
    n = len(data)  # Sample size
    mean = np.mean(data)  # Sample mean
    std_err = stats.sem(data)  # Standard error of the mean

    # Calculate the critical value from the t-distribution
    critical_value = stats.t.ppf((1 + confidence) / 2, n - 1)

    # Calculate the margin of error
    margin_of_error = critical_value * std_err

    return margin_of_error

# Example usage
if __name__ == "__main__":
    # Simulate sample data
    np.random.seed(42)  # For reproducibility
    sample_data = np.random.normal(loc=50, scale=10, size=100)  # Sample of size 100

    # Calculate margin of error for 95% confidence level
    me = margin_of_error(sample_data, confidence=0.95)

    print(f"Margin of Error at 95% confidence level: {me:.2f}")
```

**11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.**

---



**Bayesian Inference Using Bayes' Theorem:**
Bayes' Theorem provides a way to update the probability of a hypothesis based on new evidence. The theorem is mathematically expressed as:

[ P(H|D) = \frac{P(D|H) \cdot P(H)}{P(D)} ]

Where:

 - ( P(H|D) ) is the posterior probability: the probability of the hypothesis (H ) given the data ( D ).
 - ( P(D|H) ) is the likelihood: the probability of observing the data ( D ) given that ( H ) is true.
 - ( P(H) ) is the prior probability: the initial probability of the hypothesis before observing the data.
 - ( P(D) ) is the evidence: the total probability of observing the data under all possible hypotheses.

 - Implementation in Python:
    Determine the probability of having a disease based on a positive test result.




```python
import numpy as np

def bayesian_inference(prior, likelihood, false_positive_rate):
    """
    Calculate the posterior probability using Bayes' Theorem.

    Parameters:
    prior (float): Prior probability of the hypothesis (e.g., disease prevalence)
    likelihood (float): Likelihood of a positive test given the disease
    false_positive_rate (float): Probability of a positive test given no disease

    Returns:
    float: Posterior probability
    """
    # Calculate the evidence (total probability of a positive test)
    evidence = (likelihood * prior) + (false_positive_rate * (1 - prior))

    # Calculate the posterior probability
    posterior = (likelihood * prior) / evidence
    return posterior

# Example parameters
prior = 0.01  # Prior probability (e.g., 1% prevalence of the disease)
likelihood = 0.9  # Likelihood of a positive test given the disease (90%)
false_positive_rate = 0.05  # Probability of a positive test given no disease (5%)

# Calculate posterior probability
posterior = bayesian_inference(prior, likelihood, false_positive_rate)

print(f"Posterior Probability: {posterior:.4f}")
```

**Explanation of the Process:**

Bayesian inference updates the probability of a hypothesis using Bayes' Theorem:

1. **Define the Hypothesis**: Identify the hypothesis \( H \) (e.g., a patient has a disease).
2. **Prior Probability**: Set the initial belief \( P(H) \) (e.g., disease prevalence).
3. **Likelihood**: Determine \( P(D|H) \), the probability of observing data \( D \) if \( H \) is true (e.g., positive test result).
4. **False Positive Rate**: Assess \( P(D|\neg H) \), the probability of a positive test when the disease is absent.
5. **Evidence**: Calculate total probability \( P(D) \):
   \[
   P(D) = P(D|H) \cdot P(H) + P(D|\neg H) \cdot P(\neg H)
   \]
6. **Posterior Probability**: Apply Bayes' Theorem:
   \[
   P(H|D) = \frac{P(D|H) \cdot P(H)}{P(D)}
   \]
7. **Interpret Results**: The posterior \( P(H|D) \) reflects the updated belief after considering the evidence.

Example:
In the example, we calculated the probability of a patient having a disease after a positive test result, updating our belief based on prior prevalence and test accuracy.




**12.Perform a Chi-square test for independence between two categorical variables in Python.**

---



```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Sample data: Contingency table
# Let's say we have two categorical variables: Gender (Male, Female) and Preference (A, B)
data = np.array([[30, 10],  # Male preferences
                 [20, 40]]) # Female preferences

# Create a DataFrame for better visualization (optional)
df = pd.DataFrame(data, columns=['Preference A', 'Preference B'], index=['Male', 'Female'])
print("Contingency Table:")
print(df)

# Perform Chi-square test
chi2_stat, p_value, dof, expected = chi2_contingency(data)

# Output the results
print(f"\nChi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)
```

**Interpretation of Results**

- **Chi-square Statistic**: Indicates the magnitude of the difference between observed and expected frequencies; a higher value suggests a stronger association between the variables.
  
- **P-value**: If the p-value is less than the significance level (e.g., 0.05), we reject the null hypothesis, indicating a significant association between the two categorical variables.

- **Degrees of Freedom**: Calculated as \((\text{number of rows} - 1) \times (\text{number of columns} - 1)\); used to determine the critical value for the Chi-square distribution.

- **Expected Frequencies**: These represent the frequencies we would expect if there were no association between the variables; they help assess the validity of the null hypothesis.


**13.Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.**

---



```python
import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for a Chi-square test based on observed data.

    Parameters:
    observed (array-like): Observed frequency table

    Returns:
    array: Expected frequencies
    """
    # Convert observed data to a NumPy array
    observed = np.array(observed)

    # Calculate the total number of observations
    total = np.sum(observed)

    # Calculate row and column totals
    row_totals = np.sum(observed, axis=1)
    col_totals = np.sum(observed, axis=0)

    # Calculate expected frequencies
    expected = np.outer(row_totals, col_totals) / total
    return expected

# Example observed data: Contingency table
observed_data = np.array([[30, 10],  # Group 1
                          [20, 40]]) # Group 2

# Calculate expected frequencies
expected_frequencies = calculate_expected_frequencies(observed_data)

# Output the results
print("Observed Frequencies:")
print(observed_data)
print("\nExpected Frequencies:")
print(expected_frequencies)
```

**14.Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.**

---



```python
import numpy as np
from scipy.stats import chisquare

# Example observed data (e.g., counts of outcomes)
observed_data = np.array([50, 30, 20])  # Observed frequencies for three categories

# Example expected distribution (e.g., expected proportions)
# Let's say we expect a uniform distribution across the three categories
total_observations = np.sum(observed_data)
expected_proportions = np.array([1/3, 1/3, 1/3])  # Uniform distribution
expected_data = expected_proportions * total_observations  # Calculate expected frequencies

# Perform the Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(observed_data, expected_data)

# Output the results
print(f"Observed Frequencies: {observed_data}")
print(f"Expected Frequencies: {expected_data}")
print(f"\nChi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The observed data does not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed data fits the expected distribution.")
```

**16.Implement an F-test using Python to compare the variances of two random samples.**

---



```python
import numpy as np
from scipy.stats import f

# Sample data
sample1 = np.array([10, 12, 23, 23, 16, 23, 21, 16])
sample2 = np.array([14, 15, 19, 20, 22, 20, 25, 26])

# Sample sizes
n1 = len(sample1)
n2 = len(sample2)

# Sample variances
var1 = np.var(sample1, ddof=1)  # Use ddof=1 for sample variance
var2 = np.var(sample2, ddof=1)

# F-statistic (larger variance on top)
F = var1 / var2 if var1 > var2 else var2 / var1

# Degrees of freedom
df1 = n1 - 1
df2 = n2 - 1

# p-value (two-tailed)
p_value = 2 * min(
    f.cdf(F, df1, df2),  # Left tail
    1 - f.cdf(F, df1, df2)  # Right tail
)

print("Variance 1:", var1)
print("Variance 2:", var2)
print("F-statistic:", F)
print("p-value:", p_value)

# Conclusion
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: Variances are significantly different.")
else:
    print("Fail to reject null hypothesis: No significant difference in variances.")
```

**17.Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results**

---



```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data: three groups with different means
group1 = np.array([20, 22, 19, 24, 30])
group2 = np.array([30, 32, 29, 35, 40])
group3 = np.array([25, 27, 22, 30, 28])

# Perform one-way ANOVA
F_statistic, p_value = stats.f_oneway(group1, group2, group3)

# Print the results
print(f"F-statistic: {F_statistic:.4f}")
print(f"p-value: {p_value:.4f}")

# Interpretation of the results
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis: at least one group mean is significantly different.")
else:
    print("Fail to reject the null hypothesis: no significant difference between group means.")

# Plotting the results
data = [group1, group2, group3]
plt.figure(figsize=(8, 6))
sns.boxplot(data=data)
plt.xticks(ticks=[0, 1, 2], labels=['Group 1', 'Group 2', 'Group 3'])
plt.title('Boxplot of Groups')
plt.ylabel('Values')
plt.xlabel('Groups')
plt.grid(axis='y')
plt.show()
```

Interpretation of the Results

- **F-statistic**: 8.4292
  - This value indicates a strong ratio of variance between the groups compared to the variance within the groups, suggesting that the differences in group means are substantial.

- **p-value**: 0.0052
  - This p-value is significantly less than the common alpha level of 0.05, indicating strong evidence against the null hypothesis.

Conclusion:
- **Reject the null hypothesis**: Since the p-value (0.0052) is less than 0.05, we reject the null hypothesis.

**19.Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA**

---



```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions(groups):
    """
    Check the assumptions for one-way ANOVA: normality, independence, and equal variance.

    Parameters:
    groups (list of arrays): List containing the groups of data to be tested.

    Returns:
    None
    """

    # Check Normality using Shapiro-Wilk test
    print("Normality Test Results:")
    for i, group in enumerate(groups):
        stat, p_value = stats.shapiro(group)
        print(f"Group {i+1}: W-statistic = {stat:.4f}, p-value = {p_value:.4f}")
        if p_value < 0.05:
            print("  -> Reject the null hypothesis: The group is not normally distributed.")
        else:
            print("  -> Fail to reject the null hypothesis: The group is normally distributed.")

    # Check Equal Variance using Levene's test
    stat, p_value = stats.levene(*groups)
    print("\nEqual Variance Test (Levene's Test):")
    print(f"Levene's W-statistic = {stat:.4f}, p-value = {p_value:.4f}")
    if p_value < 0.05:
        print("  -> Reject the null hypothesis: The variances are not equal.")
    else:
        print("  -> Fail to reject the null hypothesis: The variances are equal.")

    # Visualize the data distribution
    plt.figure(figsize=(12, 6))

    # Boxplot
    plt.subplot(1, 2, 1)
    sns.boxplot(data=groups)
    plt.title('Boxplot of Groups')
    plt.xlabel('Groups')
    plt.ylabel('Values')

    # Q-Q Plots for Normality
    plt.subplot(1, 2, 2)
    for i, group in enumerate(groups):
        stats.probplot(group, dist="norm", plot=plt)
    plt.title('Q-Q Plots for Normality')
    plt.xlabel('Theoretical Quantiles')
    plt.ylabel('Sample Quantiles')

    plt.tight_layout()
    plt.show()

# Example usage
group1 = np.array([20, 22, 19, 24, 30])
group2 = np.array([30, 32, 29, 35, 40])
group3 = np.array([25, 27, 22, 30, 28])

check_anova_assumptions([group1, group2, group3])
```

**20.Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results**

---



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.graphics.factorplots import interaction_plot

# 1. Generate Sample Data
np.random.seed(0)
df = pd.DataFrame({
    'TeachingMethod': np.repeat(['Traditional', 'Online'], 30),
    'StudentLevel': np.tile(np.repeat(['Beginner', 'Intermediate', 'Advanced'], 10), 2),
    'TestScore': np.concatenate([
        np.random.normal(70, 5, 10),  # Traditional - Beginner
        np.random.normal(75, 5, 10),  # Traditional - Intermediate
        np.random.normal(80, 5, 10),  # Traditional - Advanced
        np.random.normal(65, 5, 10),  # Online - Beginner
        np.random.normal(70, 5, 10),  # Online - Intermediate
        np.random.normal(78, 5, 10),  # Online - Advanced
    ])
})

# 2. Run Two-Way ANOVA
model = ols('TestScore ~ C(TeachingMethod) * C(StudentLevel)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print("\nTwo-Way ANOVA Results:\n")
print(anova_table)

# 3. Interaction Plot
plt.figure(figsize=(8,6))
interaction_plot(df['StudentLevel'], df['TeachingMethod'], df['TestScore'],
                 markers=['o', 's'], colors=['blue', 'green'], ms=10)
plt.title('Interaction Plot: Teaching Method vs Student Level on Test Score')
plt.ylabel('Mean Test Score')
plt.xlabel('Student Level')
plt.grid(True)
plt.show()

# 4. Boxplot Visualization
plt.figure(figsize=(8,6))
sns.boxplot(x='StudentLevel', y='TestScore', hue='TeachingMethod', data=df)
plt.title('Boxplot: Test Score by Teaching Method and Student Level')
plt.ylabel('Test Score')
plt.xlabel('Student Level')
plt.legend(title='Teaching Method')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Set seed for reproducibility
np.random.seed(42)

# 1️⃣ Simulate one sample of data from a normal distribution
sample = np.random.normal(loc=50, scale=5, size=30)  # mean=50, std=5, n=30

# Plot the sample
sns.histplot(sample, kde=True)
plt.title("Histogram of Simulated Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# 2️⃣ One-sample t-test: is sample mean different from 52?
t_stat, p_value = stats.ttest_1samp(sample, popmean=52)
print("\nOne-Sample T-Test:")
print(f"T-statistic = {t_stat:.3f}, p-value = {p_value:.4f}")

# Interpret result
if p_value < 0.05:
    print("✅ Significant: Sample mean is different from 52.")
else:
    print("❌ Not significant: No strong evidence sample mean differs from 52.")

# 3️⃣ Simulate a second sample
sample2 = np.random.normal(loc=53, scale=5, size=30)  # mean=53, std=5, n=30

# 4️⃣ Two-sample t-test (independent samples)
t_stat_2samp, p_val_2samp = stats.ttest_ind(sample, sample2)
print("\nTwo-Sample T-Test:")
print(f"T-statistic = {t_stat_2samp:.3f}, p-value = {p_val_2samp:.4f}")

if p_val_2samp < 0.05:
    print("✅ Significant: Means of two samples are different.")
else:
    print("❌ Not significant: No strong evidence of difference in means.")
```

**24.Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results**

---



```python
import numpy as np
import scipy.stats as stats

# Parameters
np.random.seed(42)  # For reproducibility
n = 30  # Sample size
sigma_0 = 10  # Hypothesized population standard deviation
sigma_0_squared = sigma_0 ** 2  # Hypothesized population variance

# Simulate random data from a normal distribution
data = np.random.normal(loc=50, scale=sigma_0, size=n)

# Calculate sample variance
sample_variance = np.var(data, ddof=1)  # ddof=1 for sample variance

# Chi-square test statistic
chi_square_statistic = (n - 1) * sample_variance / sigma_0_squared

# Degrees of freedom
df = n - 1

# Calculate p-value for two-tailed test
p_value = 2 * (1 - stats.chi2.cdf(chi_square_statistic, df))

# Print results
print(f"Sample Variance: {sample_variance:.4f}")
print(f"Chi-square Statistic: {chi_square_statistic:.4f}")
print(f"Degrees of Freedom: {df}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The population variance is significantly different from the hypothesized value.")
else:
    print("Fail to reject the null hypothesis: No significant difference in population variance.")
```

RESULTS:
- If the p-value is less than the significance level (α = 0.05), we reject the null hypothesis, concluding that the population variance is significantly different from the hypothesized value.

- If the p-value is greater than or equal to α, we fail to reject the null hypothesis, indicating no significant difference in population variance.

**26.Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.**

---



```python
import numpy as np
from scipy.stats import f
import matplotlib.pyplot as plt

# Generate two sample datasets
np.random.seed(0)
data1 = np.random.normal(50, 10, 100)  # Mean=50, SD=10, n=100
data2 = np.random.normal(50, 15, 100)  # Mean=50, SD=15, n=100

# Calculate sample variances
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

# Determine the F-statistic
if var1 > var2:
    F = var1 / var2
    dfn, dfd = len(data1) - 1, len(data2) - 1
else:
    F = var2 / var1
    dfn, dfd = len(data2) - 1, len(data1) - 1

# Compute p-value (two-tailed)
p_value = 2 * min(f.cdf(F, dfn, dfd), 1 - f.cdf(F, dfn, dfd))

# Significance level
alpha = 0.05

# Output results
print(f"Variance1: {var1:.2f}, Variance2: {var2:.2f}")
print(f"F-statistic: {F:.4f}")
print(f"Degrees of freedom: {dfn}, {dfd}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")
```

**27.Perform a Chi-square test for goodness of fit with simulated data and analyze the results.**

---



```python

import numpy as np
from scipy.stats import chisquare
import matplotlib.pyplot as plt

# Simulate categorical data (observed frequencies)
np.random.seed(42)
observed = np.random.choice(['A', 'B', 'C', 'D'], size=200, p=[0.2, 0.3, 0.3, 0.2])
unique, counts = np.unique(observed, return_counts=True)
observed_freq = dict(zip(unique, counts))

# Expected frequencies under uniform distribution
total = sum(counts)
expected_freq = {key: total / len(observed_freq) for key in observed_freq}

# Convert to ordered arrays
categories = sorted(observed_freq.keys())
obs = np.array([observed_freq[cat] for cat in categories])
exp = np.array([expected_freq[cat] for cat in categories])

# Perform Chi-square test
chi2_stat, p_val = chisquare(f_obs=obs, f_exp=exp)

# Output
print("Observed Frequencies:", obs)
print("Expected Frequencies:", exp)
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_val:.4f}")

if p_val < 0.05:
    print("Reject H₀: The data does not follow the expected distribution.")
else:
    print("Fail to reject H₀: No significant difference from expected distribution.")

```

Visualization:

```python
x = np.arange(len(categories))
bar_width = 0.35

plt.bar(x - bar_width/2, obs, width=bar_width, label='Observed', color='skyblue')
plt.bar(x + bar_width/2, exp, width=bar_width, label='Expected', color='salmon')

plt.xticks(x, categories)
plt.ylabel("Frequency")
plt.title("Observed vs Expected Frequencies")
plt.legend()
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Results:
- Since the p-value (0.3855) > 0.05, we fail to reject the null hypothesis.

- There is no statistically significant difference between the observed and expected frequencies. The simulated data reasonably fits a uniform distribution across categories A, B, C, and D.
